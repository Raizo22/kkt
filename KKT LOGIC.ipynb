```python
from sympy import*
from IPython.display import Math,display
def kkt():
  m=int(input("Berapa variabel yang digunakan : "))
  var=[]#variabel yang digunakan
  for i in range(m):
    simbol=input(f"Masukan variabel ke-{i+1} : ")
    var.append(simbol)
  var=symbols(var)
  fungsi=input("Masukan fungsi objektif : ")
  fungsi=sympify(fungsi)
  n=int(input("Ada berapa kendala : "))
  G=0#jumlah pertidaksamaan
  H=0#jumlah persamaan
  g=[]#kendala tanpa var
  h=[]#kendala tanpa var
  c_g=[]#pertidaksamaan
  c_h=[]#persamaan
  kendala=[]#fungsi kendala
  tanda=[]
  var_g=[]
  var_h=[]
  for i in range(n):
    func=input(f"Masukan fungsi kendala ke-{i+1} : ")
    kendala.append(sympify(func))
    tan=input("Masukan tanda : ")
    tanda.append(tan)
    if tanda[i] == '<=':
      G+=1
      c_g.append(kendala[i])
      g.append(kendala[i])
      var_g.append(symbols(f'lambda{i+1}'))
    elif tanda[i] == '>=':
      G+=1
      c_g.append(-kendala[i])
      g.append(-kendala[i])
      var_g.append(symbols(f'lambda{i+1}'))
    else:
      H+=1
      c_h.append(kendala[i])
      h.append(kendala[i])
      var_h.append(symbols(f'pi{i+1}'))
  hesian=hessian(fungsi,var)
  kondisi=hesian.is_positive_definite
  c_g=[c_g[i]*var_g[i] for i in range (G)]
  c_h=[c_h[i]*var_h[i] for i in range (H)]
  if kondisi==True:
    if H==0:
      h=zeros(1,1)
      c_h=zeros(len(var),1)
      zero_h=zeros(len(var),1)
      zero_h+=c_h
    else:
      c_h=Matrix(c_h)
      h=Matrix(h)
      gradh=[diff(i,j) for i in c_h for j in var]
      gradh=Matrix(gradh).reshape(H,len(var)).T
      zero_h=zeros(len(var),1)
      for i in range(len(gradh[0,:])):
        zero_h+=gradh[:,i]
    if G==0:
      g=zeros(1,1)
      c_g=zeros(len(var),1)
      zero_g=zeros(len(var),1)
      zero_g+=c_g
    else:
      c_g=Matrix(c_g)
      g=Matrix(g)
      gradg=[diff(i,j) for i in c_g for j in var]
      gradg=Matrix(gradg).reshape(G,len(var)).T
      zero_g=zeros(len(var),1)
      for i in range(len(gradg[0,:])):
        zero_g+=gradg[:,i]
    gradf=[diff(fungsi,j)for j in var]
    gradf=Matrix(gradf)
    L=zero_g+zero_h+gradf
    L=L.col_join(c_g)
    L=L.col_join(h)
    L=L.col_join(g)
    gabung=var_g+var_h+var
    solusi=solve(L,gabung)
    solutions=[dict(zip(gabung,i))for i in solusi]
    return fungsi,solutions,g,h
  else:
    print("Solusi tidak dapat ditemukan karena definit negatif")
    return fungsi


```
